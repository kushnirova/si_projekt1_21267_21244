## Przygotowanie datasetu

```python
import pandas as pd
import numpy as np

# tworzenie losowych danych
dataset = np.random.rand(1000, 3)
dataset = pd.DataFrame(dataset, columns=['Czas [tg]', 'Koszt [zł]', 'Przychód [zł]'])

# mapowanie czasu na tygodnie (zakres od 0 do 52)
dataset['Czas [tg]'] = (dataset['Czas [tg]'] * 52).astype(int)

# mapowanie kosztów na skalę 0-1000000 zł
dataset['Koszt [zł]'] = (dataset['Koszt [zł]'] * 1000).astype(int) * 1000

# mapowanie przychodów na skalę 0-1000000 zł
dataset['Przychód [zł]'] = (dataset['Przychód [zł]'] * 1000).astype(int) * 1000 + dataset['Koszt [zł]']

print(display(dataset.head(10)))
```
